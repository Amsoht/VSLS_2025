# Setup

```python
# package imports
import pandas as pd
from vsls_2025.custom_fit_func import higher
```

```python
# load and show parameters from csv file
parameters_df = pd.read_csv("../results/fitting_parameters.csv")
parameters_df
```

# Example of calculating fit data


```python
# load or generate desired timestamps where you want to calculate the fit data
# usually in python, you will have your time data as datetime objects
timestamps = pd.date_range(start="2025-06-01", end="2025-06-10", freq="D")

# the fit function however needs numerical values, i.e. fractional years, so we need to convert the timestamps to fractional years
from IAU_baseline.baseline import datetime_to_fractionalyear 
timestamps_fractional_years = datetime_to_fractionalyear(timestamps)
timestamps_fractional_years
```

```python
# calculate CH2Cl2 data for these timestamps using the fit parameters
ch2cl2_fit_data = higher(timestamps_fractional_years, *parameters_df["CH2Cl2"].values[:-1])
ch2cl2_fit_data
```
